```python
import pandas as pd
from functools import reduce
```

```python
# Read and rename columns for each dataset
df1 = pd.read_csv("data/BOPTEXP.csv")
df1.rename(columns={'observation_date': 'date', 'BOPTEXP': 'exports_of_gs'}, inplace=True)

df2 = pd.read_csv("data/BOPTIMP.csv")
df2.rename(columns={'observation_date': 'date', 'BOPTIMP': 'imports_of_gs'}, inplace=True)

df3 = pd.read_csv("data/brent_spot_prices.csv")
df3.rename(columns={'Date': 'date'}, inplace=True)

df4 = pd.read_csv("data/BUSINV.csv")
df4.rename(columns={'observation_date': 'date', 'BUSINV': 'total_business_inventories'}, inplace=True)

df5 = pd.read_csv("data/CE16OV.csv")
df5.rename(columns={'observation_date': 'date', 'CE16OV': 'civilian_employment'}, inplace=True)

df6 = pd.read_csv("data/CLF16OV.csv")
df6.rename(columns={'observation_date': 'date', 'CLF16OV': 'civilian_labor_force'}, inplace=True)

df7 = pd.read_csv("data/CPIAUCSL.csv")
df7.rename(columns={'observation_date': 'date', 'CPIAUCSL': 'consumer_price_index'}, inplace=True)

df8 = pd.read_csv("data/DFF.csv")
df8.rename(columns={'observation_date': 'date', 'DFF': 'federal_funds_rate'}, inplace=True)

df9 = pd.read_csv("data/DPCERAM1M225NBEA.csv")
df9.rename(columns={'observation_date': 'date', 'DPCERAM1M225NBEA': 'real_personal_consumption_expenditures'}, inplace=True)

df10 = pd.read_csv("data/OPEC_crude_oil_capacity.csv")
df10.rename(columns={'period': 'date', 'OPEC_crude_oil_capacity': 'opec_crude_oil_capacity'}, inplace=True)

df11 = pd.read_csv("data/OPEC_crude_oil_production.csv")
df11.rename(columns={'period': 'date', 'OPEC_crude_oil_production': 'opec_crude_oil_production'}, inplace=True)

df12 = pd.read_csv("data/PPIACO.csv")
df12.rename(columns={'observation_date': 'date', 'PPIACO': 'producer_price_index'}, inplace=True)

df13 = pd.read_csv("data/RPI_PC1.csv")
df13.rename(columns={'observation_date': 'date', 'RPI_PC1': 'real_personal_income'}, inplace=True)

df14 = pd.read_csv("data/SP500_dividend_yield.csv")
df14.rename(columns={'Date': 'date', 'Dividend_Yield': 'sp500_dividend_yield'}, inplace=True)

df15 = pd.read_csv("data/SP500_PE_ratio.csv")
df15.rename(columns={'Date': 'date', 'PE_ratio': 'sp500_pe_ratio'}, inplace=True)

df16 = pd.read_csv("data/TOTCI.csv")
df16.rename(columns={'observation_date': 'date', 'TOTCI': 'commercial_and_industrial_loans'}, inplace=True)

df17 = pd.read_csv("data/UMCSENT.csv")
df17.rename(columns={'observation_date': 'date', 'UMCSENT': 'consumer_sentiment_index'}, inplace=True)

df18 = pd.read_csv("data/UNRATE.csv")
df18.rename(columns={'observation_date': 'date', 'UNRATE': 'unemployment_rate'}, inplace=True)

df19 = pd.read_csv("data/USD_index.csv")
df19.rename(columns={'US_dollar_index': 'usd_index'}, inplace=True)

df20 = pd.read_csv("data/USREC.csv")
df20.rename(columns={'observation_date': 'date', 'USREC': 'us_recession_indicator'}, inplace=True)

df21 = pd.read_csv("data/VIX_index.csv")
df21.rename(columns={'VIX': 'vix_index'}, inplace=True)
```

```python
dfs = [
    df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, 
    df11, df12, df13, df14, df15, df16, df17, df18, df19, df20, df21
]

# Convert 'date' columns to datetime
for df in dfs:
    df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Merge all dataframes on 'date'
df_merged = reduce(lambda left, right: pd.merge(left, right, on='date', how='inner'), dfs)

# Sort by date
df_merged = df_merged.sort_values('date').reset_index(drop=True)

print(df_merged.head)
print(len(df_merged))
```

```python
# Save the merged dataset to CSV
df_merged.to_csv("data/merged_datasets.csv", index=False)
```
